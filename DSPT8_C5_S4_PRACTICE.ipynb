Task 1

Information Resources publishes data on market share for various products. One of their survey reports shares that Oreo controls about 15% of the market for cookie brands. Let's say 25 purchasers of cookies are selected randonly from the population.

```python
n=25
p=0.15
q=1-p
x=np.arange(26)
```

```python
# a. What is the probability that fewer than five purchasers choose Oreo?

a=st.binom.cdf(4,n,p)
print('Probability that fewer than five purchasers choose Oreo is ',round(a,2))
```

```python
# b What is the probability of more than ten purchasers choosing Oreo?

b=st.binom.cdf(10,n,p)
b=1-b
print('Probability of more than ten purchasers choosing Oreo is ',b)
```

```python
# c What is the mean and the standard deviation of those purchases?

mean=n*p
std=np.sqrt(n*p*q)

print('Mean of purchases : {}\nStandard Deviation of purchases : {}'.format(mean,round(std,2)))
```

```python
# d Plot the distribution for those purchases using the Python function.

plt.plot(x,st.binom.pmf(x,n,p),'ro--')
plt.xlabel('Number of purchasers choosing oreo')
plt.ylabel('Probability of purchasers choosing oreo')
plt.title('Binomial');
```

Probability is highest for 3 purchasers choosing oreo and it zero after 10.

Task 2 :

Adriana, a manager of Rockin's Burger restaurant, is interested in taking a more statistical approach to predict the number of customers on weekend nights. She began gathering data by assigning a host to count customers every five minutes from6 p.m to 7 p. m every Saturday for three consecutive weeks (data shown on the next slide).

Help the manager calculate the probabilities in parts (a) through (e) for any given five-minute interval between 6 p.m and 7 pm on Saturday.

```python
k=np.arange(0,16) 

rate_5=3.5
poi_prob_1=st.poisson.pmf(k,rate_5)

rate_10=rate_5*2
poi_prob_2=st.poisson.pmf(k,rate_10)

rate_15=rate_5*3
poi_prob_3=st.poisson.pmf(k,rate_15)
```

```python
#a What is the probability that no customers arrive in five- minute interval of time?

a=poi_prob_1[0]
print('Probability that no customers arrive in five-minute interval of time : ',a)
```

```python
#b What is the probability that six or more customers arrive in five-minute interval of time?

b=sum(poi_prob_1[6:])
print('Probability that no customers arrive in five-minute interval of time : ',b)
```

```python
#c What is the probability that fewer than four customers arrive in a 10-minute interval of time?

c=st.poisson.cdf(3,rate_10)
print('Probability that fewer than four customers arrive in a 10-minute interval of time: ',c)
```

```python
#d What is the probability that between three and six (inclusive) customers arrive in 10-minute of time?

d=sum(poi_prob_2[3:7])
print('Probability that between three and six (inclusive) customers arrive in 10-minute of time : ',d)
```

```python
#e What is the probability that exactly eight customers arrive in a 15-minute interval of time? 

e=poi_prob_3[8]
print('probability that exactly eight customers arrive in a 15-minute interval of time : ',e)
```

```python
#f Sketch the graphs of all the above Poisson distributions.

plt.plot(k,poi_prob_1,'ro--')
plt.xlabel('Number of custumers')
plt.ylabel('Probability of customers arrive in five-minute interval of time')
plt.title('Binomial');
```

From graph,probability of three customers arrive in a 5-minute interval of time is highest and after 10 probability is zero

From graph,probability of six customers arrive in a 10-minute interval of time is highest and after 14 probability is almost zero
