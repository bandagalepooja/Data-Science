Task 11: Find the average post-test scores of male students for suburban and urban locations.Find the relative percentage for comparison

```python
a=df[(df['school_setting'].isin(['Suburban','Urban']))&(df['gender']=='Male')]   
b=a.groupby(['school_setting'])['posttest'].mean().round(2).reset_index()  #average post-test scores of male students for suburban and urban locations
print(b)
x=b.loc[0,'posttest']
y=b.loc[1,'posttest']
z=((x-y)/x)*100
print('Relative % between avg posttest scores of Suburban & Urban : ',z.round(2))       #Relative Percentage
```
