# Separating Categorical and Numerical Variables

```python
def seperate_data_types(df):
    categorical=[]
    continuous=[]
    for col in df.columns:
        if df[col].nunique()>30:
            continuous.append(col)
        else:
            categorical.append(col)
    return categorical,continuous
categorical,continuous=seperate_data_types(df)
print(tabulate({"categorical":categorical,"continuous":continuous},headers=['categorical','continuous']))
```

# User defined functions for analysis of categorical columns

This function will check for the following:

Unique values 

Mode 

Sum of all the null values 

Total count of missing values 

```python
def info_of_categorical(col):
    print(f'unique values in {col} are : {df[col].unique()}')
    print(f'Mode of {col} is : {df[col].mode()[0]}')
    print(f'Number of missing values in {col} is : {df[col].isnull().sum()}')
    if df[col].isnull().sum()==0:
        print(f'There are no null values in {col}')
```

# User defined function for calculating mean, median, standard deviation, and count of all the continuous variables 

```python
def info_of_numerical(col):
    print(f"The mean of the {col} is {df[col].mean()}")
    print(f"The median of the {col} is {df[col].median()}")
    print(f"The standard deviation of the {col} is {df[col].std()}")
    print(f"Number of missing values in the {col} is {df[col].isnull().sum()}")
    print(f"Q1 of {col} is {df[col].quantile(0.25)}")
    print(f"Q3 of {col} is {df[col].quantile(0.75)}")
```

# $$\;\;\;\; Univariate \ Analysis \ of \ Continuous \ Variables$$

```python
# Continuous Variables

continuous.remove('SK_ID_CURR')
```

```python
print(continuous)
```

```python
for i in continuous:
    info_of_numerical(i)
    print('----------------------------------------------------')
```

```python
# Missing value treatment of Continuous Variables

for i in continuous:
    mean=df[i].mean()
    df[i].fillna(mean,inplace=True)
```

```python
df.isnull().sum()     #All missing values of continuous variables are replaced by mode
```

```python
#Outliers Treatment

for col in continuous:
    mean=df[col].mean()
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    LF=Q1-(1.5*IQR)
    UF=Q3+(1.5*IQR)
    
    for i in df[col]:
        if (i<LF) or (i>UF):
            df[col].replace(to_replace=i,value=mean,inplace=True)
        else:
            pass
```

```python
#bar plot of continuous variable

continuous1=['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_GOODS_PRICE']

fig,axs=plt.subplots(nrows=3,ncols=1,figsize=(30,20))
plt.suptitle('Univariate Analysis of Continuous Feature',fontsize=35,fontweight='bold',alpha=0.8,y=1,color='red')
plt.style.use('dark_background')
plt.rcParams.update({'text.color':'white'})
axs=axs.flatten()

for i,var in enumerate(continuous1):
    sb.boxplot(x=var,data=df,ax=axs[i])
    
fig.tight_layout()
```
