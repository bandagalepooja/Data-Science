2. Frequency of the words should not be less than 40.

```python
# Import the collections library, which provides a variety of data structures, such as dictionaries and lists.
import collections

# Import the dictionary objects Dict, List, and Tuple, which are commonly used data structures in Python.
from typing import Dict,List,Tuple

"""Converts a text to a bag-of-words (BOW) representation.

  words: A list of strings representing the words in the text.
    dictionary: A dictionary mapping words to integers.

  Returns:
    A list of tuples, where each tuple contains a word ID and its frequency."""

def text2bow(words:List[str], dictionary: Dict[str, int]):

    word_frequences = collections.defaultdict(int)
    print(word_frequences)

    for word in words:
        if word not in dictionary:                           # check condition
            dictionary[word] = len(dictionary)
        word_frequences[dictionary[word]] +=1

    return list(word_frequences.items())                     # return word frequency

sample_text = text1                                           # input text
dictionary1 ={}                                               # empty dictionary
print(text2bow(sample_text.split(),dictionary1))              # calling function
print(dictionary1)
```

```python
print('Length of text1 before removing words having frequency < 40 : ',len(text1))
```

```python
# Removing words with frequency less than 40.
for k,v in zip(dictionary1.keys(),dictionary1.values()):
  if v<40:
    text1=re.sub(r'k',' ',text1)
```

```python
print('Length of text1 after removing words having frequency < 40 : ',len(text1))
```

3. Use your understanding to relate the words as a positive, negative, or neutral review.

```python
doc1 = nlp(text1)

token_without_punct1 = [token for token in doc1 if not token.is_punct] # removing punctuation

from spacy.lang.en.stop_words import STOP_WORDS  # library for stop words
filterd_sent = [] # empty list
for word in token_without_punct1: # itration over text
  if word.is_stop == False: # fetching only the text contain after removing the stop words
    filterd_sent.append(word) # appending words in empty list

from collections import Counter # importing counter
counts = Counter() # counter variable
for token in filterd_sent: # itration over filtered sentence
    counts[token.orth_] +=1 # adding counts
print(counts.most_common(85)) # printing till the count reach 40
positive_word_list = ["good", "like", "great", "love", "better", "best",                # own positive words
                      "quality", "recommend", "easy", "fun", "worth", "wonderful"]

negative_word_list = ["bad", "disappointed", "boring"]                                 # own negative words

positive_word = []
negative_word = []
neutral_word = []

for word in counts:
    if word in positive_word_list:
        positive_word.append(word)
    elif word in negative_word_list:
        negative_word.append(word)
    else:
        neutral_word.append(word)

print("positive word: \n", positive_word)
print("Negative word: \n", negative_word)
print("Neutral words: \n",neutral_word)
```
