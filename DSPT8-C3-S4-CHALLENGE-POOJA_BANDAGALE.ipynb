# SPRINT 4 : LOOPING OVER DATA STRUCTURES - CHALLENGE

MCQ Quiz

```python
Que={}
Ans={}

for i in range (0,10,1):
    Option_list=[]
    Q=input(f'Question {i+1} : ')
    for j in range (4):
        Option=input(f'Option {j+1} : ')
        Option_list.append(Option)
    Que[Q]=Option_list
    Ans[Q]=input("Enter the correct option : ")
    print('-------------------------------------------------')
    
print(Que)

print('-------------------------------------------------')

print(Ans)
```

```python
s=0
n=1
op=[1,2,3,4]

for k,v in Que.items():
    print("Q {}. {}".format(n,k))
    j=0
    for i in v:
        print('{}. {}'.format(op[j],i))
        j=j+1
    x=input('Ans : ')
  
    if x==Ans[k]:
        s=s+1
        
    n=n+1
    print('-------------------------------------------------')
    
    
print('Your Score is {}'.format(s))
```
