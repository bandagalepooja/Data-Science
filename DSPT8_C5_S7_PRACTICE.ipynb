# Hypothesis Test-II Challenge

```python
#Importing Liabrary

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
import statistics as stat
import math
```

# Task 1

```python
s=[61,93,75,67,84,95] #scores of six students
alpha=0.1
n=len(s)
dof=n-1
sam_mean=stat.mean(s)
sam_std=stat.stdev(s)
pop_mean=71
print('Sample Mean:',sam_mean)
print('Sample Std:',sam_std)
print('Sample Size:',n)
print('dof:',dof)
```

# A. What is Null & Alternate Hypothesis

H0 :
mean<71  #Average score of students is less than 71

Ha :
mean>=71 #Average score of students is greater than or equal to 71

# B. Using alpha=0.1, what is the critical value of the Hypothesis Test

```python
Tc=st.t.ppf((1-alpha),dof)
Tc
```

# C. What are the test statistics and P-value of the hypothesis test?

```python
Ts=(sam_mean-pop_mean)/(sam_std/np.sqrt(n))
Ts
```

```python
pvalue=st.t.sf(Ts,dof)
pvalue
```

# D. Draw the conclusion for the Hypothesis Test

```python
x=np.arange(-4,4,0.01)
y=st.t.pdf(x,dof)
plt.plot(x,y,label='DOF : '+str(dof))
plt.title('T Plot Distribution')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.legend();
```

```python
if Ts<Tc and pvalue>alpha:
    print("Null Hypothesis is accepted mean is less than 71")
else:
    print("Null Hypothesis is rejected and mean is greater than equal to 71")
```

# E. Discuss the Buisness Implications of your findings

The t-statistics value of 1.44 less than positive t-critical (1.47). So the decision is to accept the null hypothesis.

Average score of students is less than 71


# Task 2:

```python
sam1=[9,9,9,36,45,48,51,57,69,72,96]  #batch 1
sam2=[60,39,39,60,79,96,69,60,75,45,90]  #batch 2
x1=np.mean(sam1)
x2=np.mean(sam2)
v1=pow(stat.stdev(sam1),2)
v2=pow(stat.stdev(sam2),2)
n1=len(sam1)
n2=len(sam2)
dof=n1+n2-2
alpha=0.1
print('Sample 1 mean : ',x1)
print('Sample 1 variance : ',v1)
print('Sample 1 size : ',n1)
print('Sample 2 mean : ',x2)
print('Sample 2 variance : ',v2)
print('Sample 2 size : ',n2)
```

# A. What is Null & Alternate Hypothesis

H0 :
x1=x2 #Avg score of batch 1 & batch 2 is same

Ha :
x1<>x2 #Avg score of batch 1 & batch 2 is not same


# B.What is the critical value of the Hypothesis Test

```python
pos_Tc=st.t.ppf(1-(alpha/2),dof)
neg_Tc=st.t.ppf(alpha/2,dof)
print('Posiitive Tc : ',round(pos_Tc,2))
print('Negative Tc : ',round(neg_Tc,2))
```

# C. What are the test statistics and P-value of the hypothesis test?

```python
a=(x1-x2)
b=np.sqrt((v1*(n1-1)+v2*(n2-1))/dof)
c=np.sqrt((1/n1)+(1/n2))
Ts=a/(b*c)
print('Test Statistics (Ts) : ',round(Ts,2))
```

```python
pvalue=st.t.sf(abs(Ts),dof)*2
pvalue
```

# D. Draw the conclusion for the Hypothesis Test. Sketch the graph and state your conclusion

```python
if neg_Tc<Ts<pos_Tc and pvalue>alpha:
    print("Null Hypothesis is accepted and mean1 equal to mean2")
else:
    print("Null Hypothesis is rejected and mean1 is not equal to mean2")
```

```python
x=np.arange(-4,4,0.01)
y=st.t.pdf(x,dof)
plt.plot(x,y,label='DOF : '+str(dof))
plt.title('T Plot Distribution')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.legend();
```

# E. Discuss the Buisness Implications of your findings

The t-statistics value of -1.85 less than negative t-critical (-1.72). So the decision is to reject the null hypothesis.

There is significance difference in Average score of Batch I & Batch II.

```python
sam1=[9,9,9,36,45,48,51,57,69,72,96]   #Batch-1
sam2=[60,39,39,60,79,96,69,60,75,45,90] #Batch-2
sam3=[38,61,62,89,86,50,85,55,80,69,95] #Batch-3
sam4=[45,55,57,88,78,60,35,97,58,88,78]  #Batch-4
alpha=0.01
dfc=4-1
dfe=len(sam1)+len(sam2)+len(sam3)+len(sam4)-4
```

```python
print('dfc:',dfc)
print('dfe:',dfe)
```

# A. What is Null & Alternate Hypothesis

H0 :
mean1-mean2-mean3-mean4=0 #Average score of Batch I,Batch II,Batch III & Batch IV is same

Ha :
mean1-mean2-mean3-mean4<>0 #Average score of Batch I,Batch II,Batch III & Batch IV is not same   

# B.What is the critical value of the Hypothesis Test

```python
from scipy.stats import f,f_oneway
Fc=f.ppf(1-alpha,dfc,dfe)
print('Critical Value (Fc) : ',round(Fc,2))
```

# C. What are the test statistics and P-value of the hypothesis test?

```python
Fs,pvalue=f_oneway(sam1,sam2,sam3,sam4)
print('Test Statistics (Fs) : ',round(Fs,2))
print('Probability Value (Pv) :',round(pvalue,2))
```

# D. Draw the conclusion for the Hypothesis Test. Sketch the graph and state your conclusion

```python
rv=st.f(dfc,dfe,loc=0,scale=1)
x=np.arange(0,8,0.01)
y=rv.pdf(x)
plt.plot(x,y);
```
