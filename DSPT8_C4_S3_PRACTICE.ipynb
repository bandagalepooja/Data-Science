Task 4 : Find average income of the customers for whom loan worth of $290 or more was approved. 

```python
avg=data[(data['LoanAmount']>290) & (data['Loan_Status']=='Y')]['ApplicantIncome'].mean()
print('Average Income',round(avg,2))
```
