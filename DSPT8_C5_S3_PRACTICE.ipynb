Task 3.1.4

Stratified Sampling w.r.t engine type (vs attribute)

```python
eng_typ_grp=mtcars.groupby('vs')                 
eng_typ_name=mtcars['vs'].unique()                               
str_df=pd.DataFrame()                                
for i in eng_typ_name:                                   
    eng_data=eng_typ_grp.get_group(i)                      
    sam_data=eng_data.sample(10)
    str_df=str_df.append(sam_data)
```

```python
str_df.reset_index(drop=True,inplace=True)
str_df
```

Task 3.1.5

Clustered Sampling w.r.t engine type (vs attribute)

```python
eng_typ_grp=mtcars.groupby('vs')              
eng_typ_name=mtcars['vs'].unique()            
clus_df=pd.DataFrame()                                 
for i in eng_typ_name:                                   
    eng_data=eng_typ_grp.get_group(i)                         
    sam_data=eng_data.sample(frac=.50)                        
    clus_df=clus_df.append(sam_data)  
```

```python
clus_df.reset_index(drop=True,inplace=True)
clus_df
```

Task 3.2.1

Measure the mean, median & standard deviation of randomely selected data where sample sizes are 10,20 and 30

```python
s10=mtcars.sample(n=10)
s20=mtcars.sample(n=20)
s30=mtcars.sample(n=30)
```

```python
sam10=[s10['hp'].mean(),
s10['hp'].median(),
round(s10['hp'].std(),2)]    
sam10
```

```python
sam20=[s20['hp'].mean(),
s20['hp'].median(),
round(s20['hp'].std(),2)]    
sam20
```

```python
sam30=[round(s30['hp'].mean(),2),
s30['hp'].median(),
round(s30['hp'].std(),2)]    
sam30
```

```python
df=pd.DataFrame()
df['Measures']=['Mean','Median','Standard Deviation']
df['Sample 10']=sam10
df['Sample 20']=sam20
df['Sample 30']=sam30
df
```
