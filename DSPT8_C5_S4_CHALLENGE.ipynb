Task-1

The number of days a construction crew cannot work during summer due to weather is represented by X. The following is the probability distribution table of the same.


```python
x=np.arange(6,15,1)
P_X=[0.03,0.08,0.15,0.2,0.19,0.16,0.1,0.07,0.02]
```

a. Find the probability that a few days lost due to extreme weather conditions does not exceed 9 days the next Summer

```python
a=sum(P_X[0:4])
print('probability that a few days lost due to extreme weather conditions does not exceed 9 days the next Summer is ',a)
```

b. Find the probability that you will lose 7 to 13 days next summer.

```python
b=sum(P_X[1:8])
print('Probability that construction crew will lose 7 to 13 days next summer is ',b)
```

c. Compute the mean and the standard deviation of X. Interpret the mean in the context of the Problem


```python
mean=sum(x*P_X)
print('Mean : ',round(mean,2))

var=sum(map(lambda x,y:(x-mean)**2*y,x,P_X))
std=pow(var,0.5)
print('Standard Deviation : ',round(std,2))
```

```python
# From mean it is interpretated that for an avg of 10 days construction crew cannot work in summer due to bad weather
```

d. Find the probability that the number of days will not be lost at all next summer

```python
d=1-sum(P_X)
print('Probability that the number of days will not be lost at all next summer is ',d)
```

e. Construct the distribution plot.

```python
sb.barplot(x,P_X,palette='Purples_r')
plt.xlabel('Number of Days')
plt.ylabel('Probability that construction crew cannot work during summer')
plt.title('Distribution Plot');
```

```python
#Probability that construction crew cannot work during summer is highest for 9 number of days
```

Task-2

In the past few years, U.S. companies have used overseas outsourcing more frequently than ever. However, outsourcing is not without problems. A recent procurement survey indicates that 20% of overseas outsourcing companies use a consultant. Suppose 15 overseas outsourcing companies are randomly selected.


```python
p=0.2
q=1-p
n=15
k=np.arange(0,n+1)
prob_binom=st.binom.pmf(k,n,p)
```

a. What is the probability of exactly five overseas outsourcing companies using a consultant?

```python
a=prob_binom[5]
print('Probability of exactly five overseas outsourcing companies using a consultant : ',round(a,3))
```

b. What is the probability of more than nine overseas outsourcing companies using a consultant?


```python
b=1-st.binom.cdf(9,n,p)
print('Probability of more than nine overseas outsourcing companies using a consultant is ',b)
```

c. What is the probability that none of the Overseas outsourcing companies use a consultant?

```python
c=prob_binom[0]
print('probability that none of the Overseas outsourcing companies use a consultant is ',round(c,3))
```

d. What is the probability of four to seven (inclusive) overseas outsourcing companies using a consultant?

```python
d=sum(prob_binom[4:8])
print('Probability of four to seven (inclusive) overseas outsourcing companies using a consultant is ',round(d,3))
```

e. Construct a graph for this binomial distribution. Give your justification for the results obtained from all the previous sub-tasks using the graph.

```python
plt.plot(k,prob_binom,'go--')
plt.xlabel('Number of Companies')
plt.ylabel('Probability of companies using consultant')
plt.title('Binomial Distribution Plot');
```

```python
# Probability that more than 8 companies using consultant is nearly zero and it is higher for 3 companies
```

Task -3

Consider the average number of annual trips to amusement parks per family in the United States is 0.548 trips per year. What is the probability of randomly selecting a family and finding the following?

```python
rate1=0.548
n=20
k=np.arange(n+1)
prob_poi1=st.poisson.pmf(k,rate1)
```

a. The family did not make a trip to the amusement park last year 

```python
a=prob_poi1[0]
print('Probability that family did not make a trip to the amusement park last year is ',round(a,2))
```

b. The family took exactly one trip to the amusement park last year. 

```python
b=prob_poi1[1]
print('Probability that family took exactly one trip to the amusement park last year is ',round(b,2))
```

 C. The family took two or more trips to the amusement parkslast year

```python
c=sum(prob_poi1[2:])
print('Probability that family took two or more trips to the amusement parkslast year is ',round(c,2))
```

d. The family took three or less than three trips to amusement parks over three years

```python
rate2=rate1*3
n=20
k=np.arange(n+1)
prob_poi2=st.poisson.pmf(k,rate2)
```

```python
d=sum(prob_poi2[0:4])
print('Probability that family took three or less than three trips to amusement parks over three years is ',round(d,2))
```

e. The family took exactly four trips to amusement parks During the past six years

```python
rate3=rate1*6
n=20
k=np.arange(n+1)
prob_poi3=st.poisson.pmf(k,rate3)
```

```python
e=prob_poi3[4]
print('Probability that family took exactly four trips to amusement parks During the past six years is ',round(e,2))
```
