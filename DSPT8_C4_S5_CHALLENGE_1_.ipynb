# Data Visualization Using Matplotlib-Challenge

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('DS1_C4_S5_Car_Data_Challenge.xlsx')
col=list(df.columns)
col
```

Task 1 : 

Mr Smith wants a ranked list of cars in terms of fuel economy Which manufacturers' cars would be on this list? 

Hint 

1) Summarize manufacturer-wise city mileage data using the maximum aggregation level 

2) Use index and value properties of the summarized data to plot the bar graphs.

```python
x=df.groupby('Make')['City_Mileage_km_litre'].max().sort_values(ascending=False)
y=x.head(10)
y

plt.bar(y.index,y.values,color='blue')
plt.xlabel('Make')
plt.ylabel('City Mileage_km_litre')
plt.xticks(rotation=90)
plt.title('City Mileage Vs Make');
```

```python
#Mahindra cars have maximum city Mileage
```

Task 4 :

Help Mr Smith to compare the car mileages for the following four companies: Hyundai, Mahindra, Renault, and Skoda 

Hint

i) Filter the data for the above-mentioned manufacturers.

ii) Prepare a pivot table by calculating measures of central tendency on city and highway mileage values. 

iii) Compare all measures for these manufacturers.

```python
m=df[df['Make'].isin(['Hyundai','Mahindra','Renault','Skoda'])]                             #Data filtration
n=m.groupby('Make')['City_Mileage_km_litre','Highway_Mileage_km_litre'].agg(['mean','median']).reset_index()         
n                                                                                            #Pivottable
```

```python
n.plot(x='Make',kind='bar', stacked=False,title='Make vs Mileage',ylabel='Mileage');             #plotting bar chart
```

```python
#Renault cars have higher city mileage but Hyundai cars have highest Highway mileage amongst the four car manufacturers
```

Task 6:
    
Help Mr Smith determine which company's car offers a better spread of fuel economy irrespective of the model type. Hint

i) Calculate car mileage by averaging city and highway mileage.

ii) Summarize average car mileages by manufacturer.

iii) Sort the car mileage data in descending order.

iv) Draw the bar plot.    

```python
col=list(df.columns)
ind=col.index('Highway_Mileage_km_litre')
df.insert(ind+1,'Avg_Car_Mileage',value=0)
df['Avg_Car_Mileage']=df[['City_Mileage_km_litre','Highway_Mileage_km_litre']].mean(axis=1)
```

```python
c=df[['Make','Avg_Car_Mileage']].sort_values(by='Avg_Car_Mileage',ascending=False).dropna()
c                                                                            #Average car mileage
```

```python
d=c.groupby('Make')['Avg_Car_Mileage'].mean().sort_values(ascending=False).head(10)
d                        #Summarizing average car mileages by manufacturer & sorting car mileage in descending order
```

```python
d.plot(x=d.index,kind='bar', stacked=False,title='Make Vs Avg Car Mileage',ylabel='Avg Car Mileage',xlabel='Make',figsize=(10,6));             #Bar Chart
```

```python
#Datsun Cars have highest Car Mileage
```

Task 7 :

Help Mr Smith find the top 10 displacement trends with respect to engine capacity

Hint:

i)Summarize displacement-wise average fuel tank capacities. 

ii) Filter manufacturers who belong to the top 10 summarized fuel tank capacity values.

iii)Draw a scatter plot from the top 10 data and See if you can discern any trend?

```python
s=df.groupby('Make')['Fuel_Tank_Capacity_litre','Displacement'].mean().round(2).sort_values(by='Fuel_Tank_Capacity_litre',ascending=False).head(10).reset_index()
s                              #Summarizing displacement-wise average fuel tank capacities and filtering top 10 car with Fuel tank capacity
```

```python
r=s['Displacement'].corr(s['Fuel_Tank_Capacity_litre'])
plt.scatter(x=s['Displacement'],y=s['Fuel_Tank_Capacity_litre'],c='r',s=10,label='r='+str(r))
plt.legend()
plt.xlabel('Displacement')
plt.ylabel('Fuel_Tank_Capacity_litre')
plt.title('Displacement Vs Fuel_Tank_Capacity_litre')
plt.show()                           #Plotting scatter plot
```

```python
#There is strong positive correlation between Displacement & Fuel Tank Capacity
```
