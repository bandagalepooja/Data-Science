Task 1 : The bank officer needs to find the 'Loan_ID' and 'Gender' of customers who have an income of less than $6,000, a loan amount of more than $200,000, and the loan status as approved. Help the bank officer accomplish this task. How many such male and female customers are present?

```python
a=df[(df['ApplicantIncome']<6000) & (df['LoanAmount']>200) & (df['Loan_Status']=='Y')][['Loan_ID','Gender']]
a
```

```python
a['Gender'].value_counts()     #Count of male & female customers
```
