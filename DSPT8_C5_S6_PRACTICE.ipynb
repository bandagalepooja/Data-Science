# Hypothesis Test I Practice

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
```

# A. What is Null & Alternate Hypothesis and type of tailed test to be performed

# B. Using alpha=0.05, what is critical value of hypothesis test ?

```python
alpha=0.05
neg_zc=st.norm.ppf(alpha)
neg_zc
```

# C. What are the test statistics for hypothesis test ?

As Sample Size is given 49 which is greater than 30, population standard deviation 1.28 is given & data is normally distributed. Hence Z test can be performed 

# D. Calculate the p-value for hypothesis test.

```python
pop_mean=33.88
sam_mean=32.18
pop_std=1.28
n=49
zs=(sam_mean-pop_mean)/(pop_std/np.sqrt(n))
pvalue=st.norm.sf(abs(zs))
pvalue
```

# E. Draw the conclusion

```python
if neg_zc<zs and pvalue>alpha:
     print('Null Hypothesis Accepted')
else:
    print('Null Hypothesis Rejected')
```

# Task 2

# A. What is Null & Alternate Hypothesis and type of tailed test to be performed

# B. Using alpha=0.05, what is critical value of hypothesis test ?

```python
alpha=0.05
pos_zc=st.norm.ppf(1-alpha)
pos_zc
```

# C. What are the test statistics for hypothesis test ?

As Sample Size is given 32 which is greater than 30, population standard deviation 2.1 is given & data is normally distributed. Hence Z test can be performed 

# D. Calculate the p-value for hypothesis test.

```python
pop_mean=85
sam_mean=87.5
pop_std=2.1
n=32
zs=(sam_mean-pop_mean)/(pop_std/np.sqrt(n))
pvalue=1-st.norm.cdf(zs)
pvalue
```

# E. Draw the conclusion

```python
if zs<pos_zc and pvalue>alpha:
     print('Null Hypothesis Accepted')
else:
    print('Null Hypothesis Rejected')
```

```python
lower=pop_mean-3*pop_std
upper=pop_mean+3*pop_std
data=np.arange(lower,upper,0.1)
norm_prob=st.norm.pdf(data,loc=pop_mean,scale=pop_std)
```

```python
plt.plot(data,norm_prob,)
plt.xlabel('Yard Size')
plt.ylabel('Probability')
plt.title('Normal Distribution Plot');
```
