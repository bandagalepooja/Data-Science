Identify the manufacturer that provides a wide variety of cars by analyzing, visualizing, and comparing data using Python.

1. Group the dataset with respect to car manufacturers (Make). 

2. Calculate mean, median, standard deviation, and coefficient of variance and present them in a tabular format of engine displacement, fuel tank capacity, city mileage, highway mileage, and turning radius.

```python
#1 & #2

df1=df.groupby('Make')[['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']].agg(['mean','median','std'])
df1=df1.dropna()
df1=df1.reset_index()
df1
```

```python
df1.columns
```

```python
col=list(df1.columns)
Disind=col.index((            'Displacement',    'std'))
Capacityind=col.index(('Fuel_Tank_Capacity_litre',    'std'))
CityMileageind=col.index((   'City_Mileage_km_litre',    'std'))
HighwayMileageind=col.index(('Highway_Mileage_km_litre',    'std'))
Radiusind=col.index((  'Minimum_Turning_Radius',    'std'))

df1.insert(Disind+1,'Displacement_Cv',value=df1[(            'Displacement',    'std')]*100/df1[(            'Displacement',   'mean')])
df1.insert(Capacityind+1,'Fuel_Tank_Capacity_litre_Cv',value=df1[('Fuel_Tank_Capacity_litre',    'std')]*100/df1[('Fuel_Tank_Capacity_litre',   'mean')])
df1.insert(CityMileageind+1,'City_Mileage_Cv',value=df1[(   'City_Mileage_km_litre',    'std')]*100/df1[(   'City_Mileage_km_litre',   'mean')])
df1.insert(HighwayMileageind+1,'Highway_Mileage_Cv',value=df1[('Highway_Mileage_km_litre',    'std')]*100/df1[('Highway_Mileage_km_litre',   'mean')])
df1.insert(Radiusind+1,'Minimum_Turning_Radius_Cv',value=df1[(  'Minimum_Turning_Radius',    'std')]*100/df1[(  'Minimum_Turning_Radius',   'mean')])
```

Task 2:

Identify the factors that affect the mileage of the car and represent it visually using appropriate types of charts/graphs in Python.

1. Group the dataset with respect to the car manufacturers (Make)

2. Draw the correlation graph mileage of a car with all possible columns to identify the effect Calculate the correlation coefficient for each pair 

3. Draw a conclusion from the result of your analysis.

```python
#1

df['Car_Mileage']=(df['City_Mileage_km_litre']+df['Highway_Mileage_km_litre'])/2
```

```python
x=df[['Car_Mileage','Displacement','Make','Fuel_Tank_Capacity_litre']].dropna().groupby('Make').mean().reset_index()
x
```

```python
#2

r=x['Car_Mileage'].corr(x['Displacement'])
plt.scatter(x=x['Car_Mileage'],y=x['Displacement'],c='r',s=10,label='r='+str(r))
plt.legend()
plt.xlabel('Car_Mileage')
plt.ylabel('Displacement')
plt.title('Car_Mileage Vs Displacement')
plt.show()
```

```python
#Negative correlation between displacement and car mileage. Displacement & car mileage are inversely proportional. If displacement increases car mileage decreases
```

```python
r=x['Car_Mileage'].corr(x['Fuel_Tank_Capacity_litre'])
plt.scatter(x=x['Car_Mileage'],y=x['Fuel_Tank_Capacity_litre'],c='r',s=10,label='r='+str(r))
plt.legend()
plt.xlabel('Car_Mileage')
plt.ylabel('Fuel_Tank_Capacity_litre')
plt.title('Car_Mileage Vs Fuel_Tank_Capacity_litre')
plt.show()
```

```python
#Negative correlation between fuel tank capacity and car mileage.Fuel tank capacity & car mileage are inversely proportional. If fuel tank capacity increases car mileage decreases
```
