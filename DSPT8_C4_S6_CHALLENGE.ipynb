```python
import pandas as pd                     
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('DS1_C4_S6_House_Prices_Data_Challenge.csv')
df
```

```python
col=list(df.columns)
col.sort()
col
```

Task 1:
    
Most people prefer houses that are at least 6000 square feet and are corner houses

```python
T1=df[(df['LotArea']>=6000) & (df['LotConfig']=='Corner')]
T1
```

Task 2:
    
Many prefer houses that are one family and one story,preferably on flat land avoiding mountains or valleys or banks

```python
T2=df[(df['BldgType']=='1Fam')&(df['HouseStyle']=='1Story')&(df['LandContour']=='Lvl')]
T2
```

Task 3:
    
Compact townhouses are hot-selling in California

```python
T3=df[(df['BldgType']=='TwnhsE')|(df['BldgType']=='Twnhs')].groupby('HouseStyle')['Id'].count().reset_index()
T3
```

```python
#1 Story houses are most prefered in town houses in california
```

Task 4:
    
Houses with 3+ bedrooms, a basement, and a cement exterior are popular among customers

```python
T4=df[(df['BedroomAbvGr']>3) & (df['BsmtQual'].isnull()==False) & ((df['Exterior1st']=='CemntBd')|(df['Exterior2nd']=='CemntBd'))]
T4
```

Task 5:
    
Most families own two cars and prefer a garage with at least 500 square feet

```python
T5=df[(df['GarageArea']>=5)&(df['GarageCars']==2)]
T5
```
