Task-1

Consider that 46% of the U.S. labor force is female, of which 25% work part-time. In addition, 17.4% of all American laborers are known to be part-time workers. If a part-time worker is randomly selected, what is the probability that the worker is a female?

Let us assume size of population data is 'X'

Thus No of Females = 0.46 * X

Nos of Female Employees who work part-time = 0.25 * 0.46 * X

As 17.4 % of all American laborers are part time workers

No of part time workers = 0.174 * X

If a part-time worker is randomly selected,the probability that the worker is a female is = (0.25 * 0.46 * X)/(0.174 * X) = 0.66

Task-2

NX Partners researched to determine which changes in the office design would increase productivity. During the examination, 70% of respondents believed that reducing noise would increase productivity, and 67% said that more storage space would increase productivity. In addition, 56% of the respondents believed that reducing noise and increasing storage space would increase productivity.

A worker is randomly selected and asked about the changes in the office design. He believes that reducing noise will increase productivity. What is the probability that the worker believes that more storage space will increase productivity?

P(N)=Probability that respondent selected believed that reducing noise will increase productivity = 0.7

P(S)=Probability that respondent selected believed that more storage space would increase productivity = 0.67

P(S and N)=Probability that respondents believed that reducing noise and increasing storage space would increase productivity = 0.56

Thus P(S/N) = P(S and N)/P(N) = 0.56/0.7 = 0.8

Task 3.1.4

Stratified Sampling with respect to year and Genre

```python
Year_grp=games_sales.groupby('Year')                 
Year=games_sales['Year'].unique()                               
str_df=pd.DataFrame()                                
for i in Year:                                   
    Year_data=Year_grp.get_group(i)                      
    sam_data=Year_data.sample(5)
    str_df=str_df.append(sam_data)
str_df
```

```python
Genre_grp=games_sales.groupby('Genre')                 
Genre=games_sales['Genre'].unique()                               
str_df=pd.DataFrame()                                
for i in Genre:                                   
    Genre_data=Genre_grp.get_group(i)                      
    sam_data=Genre_data.sample(2)
    str_df=str_df.append(sam_data)
str_df
```

```python
Task 3.1.4

Clustered Sampling with respect to year and Genre
```

```python
Year_grp=games_sales.groupby('Year')                 
Year=games_sales['Year'].unique()                               
clus_df=pd.DataFrame()                                
for i in Year:                                   
    Year_data=Year_grp.get_group(i)                      
    sam_data=Year_data.sample(frac=0.04)
    clus_df=clus_df.append(sam_data)
clus_df.shape
```

```python
Genre_grp=games_sales.groupby('Genre')                 
Genre=games_sales['Genre'].unique()                               
clus_df=pd.DataFrame()                                
for i in Genre:                                   
    Genre_data=Genre_grp.get_group(i)                      
    sam_data=Genre_data.sample(frac=0.05)
    clus_df=clus_df.append(sam_data)
clus_df.shape
```

Task 3.2.1

Measure the mean, median & standard deviation of randomely selected data where sample sizes are 10,20,30 and 40

```python
s10=games_sales.sample(n=10)
s20=games_sales.sample(n=20)
s30=games_sales.sample(n=30)
s40=games_sales.sample(n=40)
```

```python
sam10=[s10['Global_Sales_M$'].mean(),
s10['Global_Sales_M$'].median(),
round(s10['Global_Sales_M$'].std(),2)]    
sam10
```

```python
sam20=[round(s20['Global_Sales_M$'].mean(),2),
s20['Global_Sales_M$'].median(),
round(s20['Global_Sales_M$'].std(),2)]    
sam20
```

```python
sam30=[round(s30['Global_Sales_M$'].mean(),2),
s30['Global_Sales_M$'].median(),
round(s30['Global_Sales_M$'].std(),2)]    
sam30
```

```python
sam40=[round(s40['Global_Sales_M$'].mean(),2),
s40['Global_Sales_M$'].median(),
round(s40['Global_Sales_M$'].std(),2)]    
sam40
```

```python
df=pd.DataFrame()
df['Measures']=['Mean','Median','Standard Deviation']
df['Sample 10']=sam10
df['Sample 20']=sam20
df['Sample 30']=sam30
df['Sample 40']=sam40
df
```
